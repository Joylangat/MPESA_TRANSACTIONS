```python
# Initialize lists to store the counts of transactions for daytime and nighttime
daytime_counts = [0] * 24
nighttime_counts = [0] * 24

# Open the CSV file
with open('datasets.csv', 'r') as csvfile:
    reader = csv.reader(csvfile)
    next(reader)  # Skip the header row

    # Iterate over each row in the CSV file
    for row in reader:
        trans_time_str = row[elements.index('TransTime')]
        trans_time = datetime.strptime(trans_time_str, "%Y%m%d%H%M%S")
        
         # Check if the transaction time is during the day or night
        if trans_time.hour >= 6 and trans_time.hour < 18:
            daytime_counts[trans_time.hour] += 1
        else:
            nighttime_counts[trans_time.hour] += 1
```

```python
# Create a list of hours for x-axis labels
hours = range(24)

# Plot the counts for daytime transactions
plt.figure(figsize=(10, 6))
plt.bar(hours, daytime_counts, label='Daytime')
plt.xlabel('Hour')
plt.ylabel('Transaction Count')
plt.title('Count of Transactions During the Day')
plt.xticks(hours)
plt.legend()
plt.show()
```

```python
# Plot the counts for nighttime transactions
plt.figure(figsize=(10, 6))
plt.bar(hours, nighttime_counts, label='Nighttime')
plt.xlabel('Hour')
plt.ylabel('Transaction Count')
plt.title('Count of Transactions During the Night')
plt.xticks(hours)
plt.legend()
plt.show()
```
